```python
import pandas as pd
import matplotlib.pyplot as plt

# =====================
# Load CSV file
# =====================
df = pd.read_csv("input_data.csv")  # change to your filename

x = df.iloc[:, 0]  # index / time
wind = df.iloc[:, 1:14]  # wind power (columns 214)
price = df.iloc[:, 14]  # price (column 15)

# =====================
# Create horizontal subplots
# =====================
fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharex=True, dpi=600)

# ---- Plot 1: Wind power ----
for i in range(wind.shape[1]):
    axes[0].plot(x, wind.iloc[:, i], label=f"Wind {i + 1}")

axes[0].set_title("Wind Power")
axes[0].set_xlabel("Index")
axes[0].set_ylabel("Power")
axes[0].legend(fontsize=8, ncol=2)

# ---- Plot 2: Price ----
axes[1].plot(x, price, label="Price")

axes[1].set_title("Price")
axes[1].set_xlabel("Index")
axes[1].set_ylabel("Price")
axes[1].legend()

# ---- Plot 3: Wind power  price ----
for i in range(wind.shape[1]):
    axes[2].plot(x, wind.iloc[:, i] * price, label=f"Wind {i + 1}  Price")

axes[2].set_title("Wind Power  Price")
axes[2].set_xlabel("Index")
axes[2].set_ylabel("Power  Price")
axes[2].legend(fontsize=8, ncol=2)

plt.tight_layout()
plt.show()
```
